```python
import tensorflow
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[
    'the cup of coffee',
    'the bottle of water',
    'she is a good girl',
    'he is a smart student',
    'I love machine learning',
    'we are learning NLP',
]
```

```python
sent
```

```python
voc_sizes=10000
```

```python
one_hot_rep=[one_hot(word,voc_sizes) for word in sent]
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
sent_len=8

emdedded_doc=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_len)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
```

```python
dim=10
```

```python


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Input

model = Sequential()
model.add(Input(shape=(sent_len,)))  
model.add(Embedding(voc_sizes, dim))

model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
model.predict(emdedded_doc)
```
